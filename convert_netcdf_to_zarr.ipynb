```python
import os
import xarray as xr
import s3fs
```

```python
netcdf_dir = 's3://dh-gapc/GrIS_committed_SLR/netcdfs/dhdynAnom/'
zarr_dir = 's3://dh-gapc/GrIS_committed_SLR/zarr/dhdynAnom/'
```

```python
s3 = s3fs.S3FileSystem(anon=False)
s3.clear_instance_cache()
netcdf_files = s3.glob(netcdf_dir + '*.nc')

for netcdf_file in netcdf_files:
    with s3.open(netcdf_file, 'rb') as f:
        zarr_file = zarr_dir + os.path.basename(netcdf_file) + '.zarr'
        if len(s3.glob(zarr_file)) > 0:
            print(zarr_file + ' already exists ... skipping')
        else:
            print(netcdf_file + ' -> ' + zarr_file)

            ds = xr.open_dataset(f, engine='h5netcdf')
            ds.to_zarr(zarr_file, consolidated=True, mode='w')
            ds.close()
    
```
